## Merging the data

```python
market = pd.read_csv('market_fact.csv')
customer = pd.read_csv('cust_dimen.csv')
product = pd.read_csv('prod_dimen.csv')
```

```python
market.head()
```

```python
customer.head()
```

```python
product.head()
```

```python
new_df = pd.merge(market, customer, on = 'Cust_id', how = 'inner')
new_df.head()
```

```python
final_df = pd.merge(new_df, product, on = 'Prod_id', how = 'inner')
final_df.head()
```

```python
final_df.to_csv('final_data.csv', index = False)
```

### Pandas Joining

```python
df1 = pd.DataFrame({'Cust_id': ['C1', 'C2', 'C3', 'C4', 'C5'],
                   'Profit': [10,12,13,41,51], 
                   'Sales': [100, 120,321,400, 500]})
```

```python
df4 = pd.DataFrame({'Ship_id': ['S1', 'S2', 'S3', 'S4'],
                   'Order_Quantity': [2,5,12,14],
                   'Cust_id' : ['C6', 'C7', 'C8', 'C9']})
```

```python
df1
```

```python
df4
```

```python
df1.join(df4, how = 'left', lsuffix = '_1', rsuffix = '_4')
```

```python
df5 = pd.DataFrame({'Ship_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
                   'Order_Quantity': [2,5,12,14, 15],
                   'Cust_id' : ['C1', 'C2', 'C3', 'C4', 'C5']})
```

```python
df5
```

```python
df1
```

```python
df1.join(df5.set_index('Cust_id'), on = 'Cust_id', how = 'left')
```

```python
pd.merge(df1, df5, on = 'Cust_id', how = 'left')
```
